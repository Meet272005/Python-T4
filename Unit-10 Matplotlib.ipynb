## Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors1=np.array([50,50,50,50,200,50,50,50,50,50,50,50,50])
plt.scatter(x,y,marker='*',c=colors1,cmap='viridis',alpha=0.5)
plt.colorbar()

x=np.array([15,7,8,7,9,5,20,9,4,13,2,19,6])
y=np.array([107,86,77,88,101,86,103,87,99,78,77,88,86])
colors=np.array(['red','green',"blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
sizes=np.array([50,50,50,50,200,50,50,50,50,50,50,50,50])
plt.scatter(x,y,c=colors,s=sizes)
plt.show()
```

## Pie chart

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([35,25,25,15])
plt.pie(y)
plt.show()

plt.figure(figsize=(10,10))
plt.title("Pie-chart",loc="center")
mylabels=["Apple","Banana","cherry","dates"]
myexplode=[0.2,0,0,0.1]
mycolor=["red","yellow","pink","brown"]
plt.pie(y,startangle=90,labels=mylabels,explode=myexplode,shadow=True,colors=mycolor,autopct="%1.1f%%")
plt.legend(loc="upper left")
# plt.tight_layout()
plt.show()
```
